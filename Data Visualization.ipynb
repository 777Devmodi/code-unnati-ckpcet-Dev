### Sub plots

```python
# defing the data form for plot 1

import numpy as np 
#importing the numpy is recommended and standard method to declare at top
#plot-1 
x=np.array([0,1,2,3,4,5])
y=np.array([10,22,15,35,55,60])

#plot-2
a=np.array([0,1,2,3,4,5])
b=np.array([5,15,22,14,6,11])

#  plt.plot(x,y,a,b)

#means 1 row 2 columns and at 1 location
plt.subplot(1,2,1)
plt.plot(x,y)
#means 1 row 2 graph and at 2 location
plt.subplot(1,2,2)
plt.plot(a,b,color="orange")
```

```python
# defing the data form for plot 1
#plot-1 
x=np.array([0,1,2,3,4,5])
y=np.array([10,22,15,35,55,60])

#plot-2
a=np.array([0,1,2,3,4,5])
b=np.array([5,15,22,14,6,11])
#spacing adding extra column
#subplot(2,2,) means 2*2 space and third parameter would be the location
#means 1 row 2 columns and at 1 location
plt.subplot(1,3,1)
plt.plot(x,y)
#means 1 row 2 graph and at 2 location
plt.subplot(1,3,3)
plt.plot(a,b,color="orange")
```

```python
# defing the data form for plot 1

# Placing Then in vertical manner 
#plot-1 
x=np.array([0,1,2,3,4,5])
y=np.array([10,22,15,35,55,60])

#plot-2
a=np.array([0,1,2,3,4,5])
b=np.array([5,15,22,14,6,11])
#spacing adding extra column
#means 1 row 2 columns and at 1 location
plt.subplot(2,1,1)
plt.plot(x,y)
#means 1 row 2 graph and at 2 location
plt.subplot(2,1,2)
plt.plot(a,b,color="orange")
```

### Histogram

```python
# it represent that in form of frequency and in range
```

```python
#random.normal will generate the normalized data
#means 170 range of 10 and 80 data
# 10 means random incremental in variation of +-10.
data=np.random.normal(170,10,80)
```

```python
print(len(data))
data
```

```python
#ploting historgram
plt.hist(data)
```

```python
#for specific bins to restrict the bar size length and divide them in categories 
#bins will divide the data into categories
plt.hist(data,bins=5)
```
